**Step 1: Create a pandas DataFrame called rides with 500 rows and the following columns:**

```python
import pandas as pd
import numpy as np

np.random.seed(13)
n_rows = 500

data = {
    "ride_id": np.arange(1, n_rows + 1),
    "driver_id": np.random.randint(1000, 1051, size=n_rows),
    "distance_km": np.round(np.random.uniform(1.0, 30.0, size=n_rows), 2),
    "fare_usd": np.round(np.random.uniform(5.0, 75.0, size=n_rows), 2),
    "ride_type": np.random.choice(["standard", "premium", "pool"], size=n_rows),
    "city": np.random.choice(["Berlin", "Seoul", "Nairobi", "Toronto", "Lima"], size=n_rows),
    "timestamp": pd.date_range(start="2025-01-01", periods=n_rows, freq="min")
}

rides = pd.DataFrame(data)

rides.head()
```

**Step 2: Save the DataFrame in three formats:**

```python
rides.to_csv("rides.csv", index=False)

rides.to_json("rides.json", orient="records", lines=True)

rides.to_parquet("rides.parquet", index=False)
```

**Step 3: Compare the file sizes of all three formats. Print the size of each file in KB and compute how many times smaller Parquet is compared to CSV.**

```python
import os

csv_size = os.path.getsize("rides.csv") / 1024
json_size = os.path.getsize("rides.json") / 1024
parquet_size = os.path.getsize("rides.parquet") / 1024

ratio = csv_size / parquet_size

print(f"CSV File size: {csv_size:.2f} KB")
print(f"JSON File size: {json_size:.2f} KB")
print(f"Parquet File size: {parquet_size:.2f} KB")
print("-" * 30)
print(f"Result: Parquet is {ratio:.2f} times smaller than CSV.")
```

* Data Format Summary
Most Faithful Type Preservation: Parquet. It is the only format that natively saves the schema metadata, ensuring that datetime and numeric types stay exactly as they were without needing manual conversion.

Smallest Size: Parquet. It uses binary columnar compression, making it significantly more compact than text-based formats. *

* When to Choose Each Format?
Parquet: Best for ML Pipelines and Big Data. Choose this for high-performance production environments where speed and storage efficiency are priorities.

CSV: Best for Portability. Choose this when you need to open the file in Excel or share it with people who don't use Python.

JSON: Best for Web & Streaming. Choose this for APIs or logs where data might be nested or needs to be read sətir-sətir (line by line). *

* 1. Which representation made the answer easiest?

 * 2. Frequent Publisher Changes?
If publisher information (like country or name) changes frequently, the Normalized (Step 2) representation is the superior choice.

Why? It follows the "Single Source of Truth" principle. You only have to update one record in the publishers table.

In the Flat and Document models, you have to find and update every single occurrence of that publisher across thousands of rows or documents, which risks data inconsistency (e.g., updating some rows but missing others). *

* 3. Accessing Books as Standalone Records?
If your primary use case is a website where users click on a book to see all its details (title, author, publisher, and all available prices), the Document (Step 3) representation is the best choice.

Why? All the data needed for that specific book is bundled together in one "document."

In a Normalized system, you would have to perform a "Join" (merging tables) every single time a page loads, which can become a performance bottleneck at scale.*
